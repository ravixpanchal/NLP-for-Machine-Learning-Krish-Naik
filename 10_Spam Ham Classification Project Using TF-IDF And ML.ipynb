```python
#data cleaning and data preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(data)):
    review=re.sub('[^a-zA-z]',' ',data['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

# BOW,TFIDF,Machine Learning Algorithms
- Preprocessing And Cleaning
- Train Test Split
- BOW And TF-IDF (Sentences--->vectors) {Preventing Data Leakage}
- Trained Our Models

```python
import pandas as pd
data=pd.read_csv('SMSSpamCollection.txt',sep="\t",names=["label","message"])
```

```python
data.head()
```

```python
## Data Cleaning And Preprocessing
import re
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(data)):
    review=re.sub('[^a-zA-z]',' ',data['message'][i])
    review=review.lower()
    review=review.split()
    review=[ps.stem(word) for word in review if not word in stopwords.words('english')]
    review=' '.join(review)
    corpus.append(review)
```

```python
corpus
```

```python
# Creating BoW
y=pd.get_dummies(data['label'])
```

```python
y=y.iloc[:,0].values
```

```python
# Train test Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(corpus,y,test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
len(X_train), len(y_train)
```

```python
X_train=cv.fit_transform(X_train).toarray()
X_test=cv.transform(X_test).toarray()
```

```python
import numpy as np
np.set_printoptions(edgeitems=30, linewidth=100000, 
    formatter=dict(float=lambda x: "%.3g" % x))
X_train
```

```python
cv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_detect_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_detect_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,classification_report
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

# Creating The TF-IDF Model

```python
#Train test split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(corpus,y,train_size=0.20)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv=TfidfVectorizer(max_features=2500,ngram_range=(1,2))
X_train=tv.fit_transform(X_train).toarray()
X_test=tv.transform(X_test).toarray()
```

```python
X_train
```

```python
tv.vocabulary_
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_tfidf_model=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred=spam_tfidf_model.predict(X_test)
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```
