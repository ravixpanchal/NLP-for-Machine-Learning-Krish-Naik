# Tokenization Example

```python
corpus=""" Hello Welcome, to Ravi Panchal's NLP tutorial.
Please do watch the entire course! to become expert in NLP.
"""
```

```python
print(corpus)
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
```

## Tokenization 


### Paragraph-> Sentences

```python
from nltk.tokenize import sent_tokenize
sent_tokenize(corpus)
```

```python
documents=sent_tokenize(corpus)
type(documents)
for sentences in documents:
    print(sentences)
```

### Paragraph->Words OR Sentences->Words

```python
from nltk.tokenize import word_tokenize
word_tokenize(corpus)
```

```python
for sentences in documents:
    print(word_tokenize(sentences))
```

```python
from nltk.tokenize import wordpunct_tokenize
wordpunct_tokenize(corpus)
```

```python
from nltk.tokenize import TreebankWordTokenizer
tokenizer=TreebankWordTokenizer()
tokenizer.tokenize(corpus)
```

# Interview Question

```python
#practise Question
from nltk.tokenize import sent_tokenize,word_tokenize


corpus="Hello, My name is Ravi. I'm from Orai, UP. I'm btech 3rd year btech Student. My college name is GSV. My father is a farmer. I'm passionate about coding."
sentences=sent_tokenize(corpus)
for sentence in sentences:
    print(sentence)

```

```python
for word in sentences :
    arr=word_tokenize(word)
    print(len(arr))
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer(r'\w+')
text="Hello, My name is Ravi. I'm from Orai, UP. I'm btech 3rd year btech Student. My college name is GSV. My father is a farmer. I'm passionate about coding."
tokens=tokenizer.tokenize(text)
print(tokens)
```
